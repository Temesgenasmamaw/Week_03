```python
# Check for missing values
missing_values_df = missing_values_table(data)
print("Missing Values in df:\n\n")
print(missing_values_df)
```

```python
# from the above NumberOfVehiclesInFleet have  1000098 null values which means all are null so drop this column
data=data.drop(columns='NumberOfVehiclesInFleet',axis=1)
```

```python
# data types
print(data.dtypes)
```

```python
# fill numeric missing value with mean
numeric_cols = data.select_dtypes(include='number').columns
data_filled_numeric = data[numeric_cols].fillna(data[numeric_cols].mean())

# Fill non-numeric columns with a placeholder or mode
non_numeric_cols = data.select_dtypes(exclude='number').columns

data_filled_non_numeric = data[non_numeric_cols].fillna('Unknown')


# Merge both numeric and non-numeric
new_data = pd.concat([data_filled_numeric, data_filled_non_numeric], axis=1)


# check missing value after hadle it
print(new_data.isnull().sum())
```

```python
# data types
print(new_data.dtypes)
```

```python
new_data.shape
```

```python
data_iqr_cleaned = remove_all_columns_outliers(new_data, method="iqr")
data_cleaned_zscore = remove_all_columns_outliers(new_data, method="zscore")
# Check the shape of the cleaned 
print(f"Original data shape: {data.shape}")
print(f"Cleaned data shape (IQR method): {data_iqr_cleaned.shape}")
print(f"Cleaned data shape (Z-score method): {data_cleaned_zscore.shape}")
```

```python
# take the z-score analysis
cleaned_data=data_cleaned_zscore.copy()
```

```python
cleaned_data.head(1)
```

```python

# Converts the TransactionMonth column from object to datetime format.

cleaned_data['TransactionMonth'] = pd.to_datetime(cleaned_data['TransactionMonth'], format='%d/%m/%Y', errors='coerce') 
```

```python
# Converts the VehicleIntroDate column from object to datetime format.

cleaned_data['VehicleIntroDate'] = pd.to_datetime(cleaned_data['VehicleIntroDate'], format='%d/%m/%Y', errors='coerce')  
```

```python
cleaned_data.to_csv('../data/cleanedDataset.csv', index=False)
```
